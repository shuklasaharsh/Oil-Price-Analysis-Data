```python
###RSP Dataset
rsp_df = pd.read_csv('RSP-Data.csv')
rsp_df.head()
```

The values under the metro cities are the RSPs of Oil for the respective cities on the particular day

```python
rsp_df['Date'] = rsp_df['Date'].astype('datetime64[ns]')
rsp_df.head()
```

```python
###Region based Unemployment dataset
unem_df = pd.read_csv('Unemployment-Data.csv')
unem_df.head()
```

```python
#Removing spaces in the names of the columns
#and deleting frequency column
unem_df = unem_df.rename({' Date': 'Date', 
                          ' Frequency': 'Frequency',  
                          ' Estimated Unemployment Rate (%)': 'Estimated Unemployment Rate (%)', 
                          ' Estimated Employed':'Estimated Employed', 
                          ' Estimated Labour Participation Rate (%)': 'Estimated Labour Participation Rate (%)'}, 
                         axis=1)
unem_df.drop('Frequency', inplace=True, axis=1)
unem_df.columns
```

```python
unem_df['Date'] = unem_df['Date'].astype('datetime64[ns]')
unem_df.head()
```

```python
#removing other states to only focus on Delhi, Mumbai, Kolkata and Chennai
Mumbai = unem_df['Region'] == "Maharashtra"
Delhi = unem_df['Region'] == "Delhi"
Kolkata = unem_df['Region'] == "West Bengal"
Chennai = unem_df['Region'] == "Tamil Nadu"

unem_df = unem_df[Mumbai | Delhi | Kolkata | Chennai]
```

```python
#splitting into 2 datasets
unem_df_rural = unem_df[unem_df['Area'] == "Rural"]
unem_df_urban = unem_df[unem_df['Area'] == "Urban"]
```

```python
#removing the area column
unem_df_urban.drop('Area', inplace=True, axis=1)
unem_df_rural.drop('Area', inplace=True, axis=1)
unem_df_urban.head()
```

```python
#merging both rural and urban datasets
unem_df_merge = pd.merge(unem_df_urban, unem_df_rural, on="Date", sort=True )
#filtering out the rows that have inconsistent values
unem_df_merge = unem_df_merge[unem_df_merge['Region_x'] == unem_df_merge['Region_y']]
#x - urban
#y - rural

#rename the columns
unem_df_merge.drop('Region_y', inplace=True, axis=1)
unem_df_merge = unem_df_merge.rename({"Estimated Unemployment Rate (%)_x": "Urban Estimated Unemployment Rate (%)", "Estimated Employed_x": "Urban Estimated Employed", "Estimated Labour Participation Rate (%)_x":"Urban Estimated Labour Participation Rate (%)"}, axis=1)
unem_df_merge = unem_df_merge.rename({"Estimated Unemployment Rate (%)_y": "Rural Estimated Unemployment Rate (%)", "Estimated Employed_y": "Rural Estimated Employed", "Estimated Labour Participation Rate (%)_y": "Rural Estimated Labour Participation Rate (%)"}, axis=1)
unem_df_merge = unem_df_merge.rename({"Region_x": "Region"}, axis=1)
unem_df_merge.head(10)
```

```python
rsp_df_delhi = rsp_df[['Date','Delhi']]
rsp_df_chennai = rsp_df[['Date','Chennai']]
rsp_df_kolkata = rsp_df[['Date','Kolkata']]
rsp_df_mumbai = rsp_df[['Date','Mumbai']]

rsp_df_delhi = rsp_df_delhi.rename({'Delhi': 'Oil RSP'}, axis=1)
rsp_df_chennai = rsp_df_chennai.rename({'Chennai': 'Oil RSP'}, axis=1)
rsp_df_kolkata = rsp_df_kolkata.rename({'Kolkata': 'Oil RSP'}, axis=1)
rsp_df_mumbai = rsp_df_mumbai.rename({'Mumbai': 'Oil RSP'}, axis=1)

#adding Region column to individual dataframes for merging more easily
rsp_df_delhi['Region'] = "Delhi"
rsp_df_chennai['Region'] = "Tamil Nadu"
rsp_df_kolkata['Region'] = "West Bengal"
rsp_df_mumbai['Region'] = "Maharashtra"
```

```python
#merging the unemployment dataframe with the oil rsp
rsp_merged_df = pd.merge(unem_df_merge, rsp_df_delhi, on=['Region','Date'], how="left")
rsp_merged_df.head()
```

```python
#merging the remaining regions' rsp values with the unemployment dataframe
new_ordered_prices = []
for row in rsp_merged_df.values:
    for entry in rsp_df_chennai.values:
        if entry[2] == row[0] and entry[0] == row[1]:
            row[8] = entry[1]
    for entry in rsp_df_kolkata.values:
        if entry[2] == row[0] and entry[0] == row[1]:
            row[8] = entry[1]
    for entry in rsp_df_mumbai.values:
        if entry[2] == row[0] and entry[0] == row[1]:
            row[8] = entry[1]
    new_ordered_prices.append(row[-1])

rsp_merged_df['Oil RSP'] = new_ordered_prices
rsp_merged_df.head()
```

```python
###Sector Based GDP
sect_df = pd.read_csv('Sector-wise-Quaterly-GDP-Inida.csv')
sect_df.head()
```

The GDP values for each sector are in crores Rupees. 

```python
#filling the NaN values with the above values
sect_df['Year/ Industry'] = sect_df['Year/ Industry'].fillna(method='ffill')
sect_df.head()
```

```python
# Q1 end - June 30th
# Q2 end - September 30th
# Q3 end - December 31st
# Q4 end - March 31st
# we choose the ending dates for each of the quarters as they can only be calculated at that point of time
```

```python
#converting the quarter ending dates into datetime format
import datetime
dates = []
for idx, row in enumerate(sect_df.values):
    #changing year for Q4
    if row[1] == "Q4":
        row[0] = sect_df.values[idx+1][0]
       
    #getting datetime for all rows
    row_date = datetime.datetime(1970,1,1)
    year = int(row[0][0:4])
    if row[1] == "Q4":
        row_date = datetime.datetime(year,3,31)
    elif row[1] == "Q1":
        row_date = datetime.datetime(year,6,30)
    elif row[1] == "Q2":
        row_date = datetime.datetime(year,9,30)
    else:
        row_date = datetime.datetime(year,12,31)
    dates.append(row_date)
```

```python
#appending datetime dates onto dataframe
sect_df['Date'] = dates
sect_df.head()
```
